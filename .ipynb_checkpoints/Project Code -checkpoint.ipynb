```python
import numpy as np
import pandas as pd
import os
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import alpaca_trade_api as tradeapi
import seaborn as sns
%matplotlib inline
import pandas_datareader.data as web
import yfinance as yf
np.random.seed(42)
```

```python
#loading data
stocks = ['AAPL', 'AMZN', 'MSFT']
data = web.DataReader(stocks, data_source = 'yahoo', start = '01/01/2019', end= '31/12/2019')['Adj Close']

#cleaning data
data.sort_index(inplace = True)
data.dropna(inplace = True)

#normalize the return dividing price at day n by price of day (n-1)
log_ret = np.log(data/data.shift(1))
log_ret.head(10)
```

```python
num_portfolios = 5000
```

```python
#saving results
return_array = np.zeros(num_portfolios)
volatility_array = np.zeros(num_portfolios)
sharpe_array = np.zeros(num_portfolios)
weight_all = np.zeros((num_portfolios, len(stocks)))
```

```python
for x in range(num_portfolios):
   # defining weight
    weight = np.array(np.random.random(3))
    weight = weight/np.sum(weight)
    #weight = [0.25, 0.5, 0.25]

    #store weights 
    weight_all[x, :] = weight
    
    #expected return 
    return_array[x] = np.sum(log_ret.mean() * weight *252)
    
    #Expected Volatility 
    volatility_array[x] = np.sqrt(np.dot(weight, np.dot(log_ret.cov() * 252, weight)))
    
    #sharpe ratio
    sharpe_array[x] = return_array[x]/volatility_array[x]
```

```python
print('Max sharpe ratio in the array; {}' .format(sharpe_array.max()))
print("Its location in the arrary: {}" .format(sharpe_array.argmax()))
```

```python
max_sr_ret = return_array[sharpe_array.argmax()]
max_sr_vol = volatility_array[sharpe_array.argmax()]

plt.figure(figsize =(12,8))
plt.scatter(volatility_array, return_array, c= sharpe_array, cmap='plasma')
plt.colorbar(label ="Sharpe Ratio")
plt.xlabel('Volatility')
plt.ylabel('Return')
plt.scatter(max_sr_vol, max_sr_ret, c='red', s=50)
plt.show()
```
